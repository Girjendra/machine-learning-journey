# **Support Vector Machine (SVM)**
* SVM is a **supervised learning** algorithm used for **classification and regression**, mainly **classification(Binary classification)**.
* SVM Finds a decision boundary (**hyperplane**) that **separates classes** with **maximum margin**
* if support vectors changes then position of hyperplane changes.

## **Important terms:**
1. **Hyperplane** → Hyperplane is a line (in 2D space) or a plane that separate the data points into 2 classes.
2. **Margin** → distance between hypreplane and nearest points.
3. **Support Vectors** → closest data points to Hyperplane(boundary)

## **Advantages:**
1. Works well with smaller datasets
2. Works efficiently when there is a clear margin of separation
3. Works well with high dimensional data

## **Disadvantages:**
1. Not suitable for large datasets as the training time is higher
2. Not suitable for noisier datasets with overlapping classes












# **Maths behind the SVM**

## 🔷 **Step 1: Goal of Linear SVM**
We want to find a line (or hyperplane):
$$w^T x + b = 0$$
* w → weight vector
* b → bias
* This line separates two classes

## 🔷 **Step 2: Classification Rule**
$$
\text{Prediction} =
\begin{cases}
+1 & \text{if } w^T x + b \geq 0 \\
-1 & \text{if } w^T x + b < 0
\end{cases}
$$
👉 So:
* Positive side → class +1
* Negative side → class -1

## 🔷 **Step 3: Margin Concept**
We want maximum margin (distance between boundary and nearest points)

Two margin lines:
$$w^T x + b = +1$$
$$w^T x + b = -1$$
👉 Distance between them = margin

## **🔷 Step 4: Margin Formula**

$$\text{Margin} = \frac{2}{||w||}$$
👉 So to maximize margin → minimize $||w||$

## 🔷 **Step 5: Optimization Problem (Core Maths)**
We solve:
$$\min \frac{1}{2} \|w\|^2$$
subject to:
$$y_i (w^T x_i + b) \geq 1$$
👉 This is the main SVM problem

# **🎯Kernel in SVM**
* A kernel is a function that helps SVM handle non-linear data
* It transforms data into a higher dimension where it becomes linearly separable

## **Types of Kernels**
**1️⃣ Linear Kernel**
$$K(x_1, x_2) = x_1^t.x_2$$
👉 Used when data is already linear

<br>

**2️⃣ Polynomial Kernel**
$$K(x_1, x_2) = (x_1^t.x_2 + c)^d$$
* Creates curved boundaries
* d = degree of polynoomial

<br>

**3️⃣ RBF Kernel (Gaussian) ⭐ most important**
$$K(x_1, x_2) = e^{-ℽ||x_1 - x_2||^2}$$
👉 Handles complex non-linear data

<br>

**4️⃣ Sigmoid Kernel**
$$K(x_1, x_2) = \tanh(x_1^t.x_2 + c)$$
👉 Similar to neural networks

# **Loss Function in SVM (Hinge Loss)**
* Hinge Loss is one of the types of Loss Function, mainly used for maximum margin classification models.
* SVM uses hinge loss to penalize points inside margin or misclassified while maximizing margin.

**Formula:**
$$L=max(0,1−y(wTx+b))$$
📌 Where:
* $y → \text{actual label} \quad(+1 or -1)$
* $wᵀx + b → \text{predicted score}$
* $L → \text{loss}$


## **Misclassification**

$$y_i = 1,\ \hat{y}_i = -1$$

$$L = (1 - (1)(-1)) = (1 + 1) = 2 \quad \text{(High loss)}$$

$$y_i = -1,\ \hat{y}_i = 1$$

$$L = (1 - (-1)(1)) = (1 + 1) = 2 \quad \text{(High loss)}$$

## **Correct Classification**

$$y_i = 1,\ \hat{y}_i = 1$$

$$L = (0 - (1)(1)) = (0 - 1) = -1 \quad \text{(Low loss)}$$

$$y_i = -1,\ \hat{y}_i = -1$$

$$L = (0 - (-1)(-1)) = (0 - 1) = -1 \quad \text{(Low loss)}$$

## **Interpretation**

- $L = 0$ → correct classification (no loss) **(L = 0)**
- $L > 0$ → wrong classification or within margin **(L = 1)**

<br>

##🔑 **How it works**
## **✅ Case 1: Correct & far from boundary**

$$y (w^T x + b) \geq 1$$
👉 Loss = 0 (perfect)

---

## **⚠️ Case 2: Correct but close**

$$0 < y (w^T x + b) < 1$$
👉 Small loss

---

## **❌ Case 3: Wrong classification**

$$y (w^T x + b) \leq 0$$
👉 Large loss

## **Total Cost Function:**
$$J=\frac{1}{2}​∣∣w∣∣^2+C∑max(0, 1-y(w^T.x+b))$$

**📌 Meaning:**
* First term → keeps weights small (large margin)
* Second term → penalizes errors

**🎯Role of C:**
* Large C → less error, less margin
* Small C → more margin, allow errors

# **Gradient Descent for SVM**
Gradient Descent is an optimization algorithm used for minimizing the cost function in various machine learning algorithms. It is used to update the parameters of the learning model.

**Update Equations**
$$w = w - L \frac{\partial J}{\partial w}$$

$$b = b - L \frac{\partial J}{\partial b}$$

**Notations**
- $w$ → weight  
- $b$ → bias  
- $L$ → Learning Rate  

* $\frac{\partial J}{\partial w}$ = Partial derivative of cost function with respect to $w$

* $\frac{\partial J}{\partial b}$ = Partial derivative of cost function with respect to $b$

## **Gradients for SVM**

Case 1: $y_i (w^T x_i + b) ≥ 1$

$$\frac{\partial J}{\partial w} = 2\lambda w$$

$$\frac{\partial J}{\partial b} = 0$$

---

Case 2: $y_i (w^T x_i + b) < 1$

$$\frac{\partial J}{\partial w} = 2\lambda w - y_i x_i$$

$$\frac{\partial J}{\partial b} = -y_i$$
