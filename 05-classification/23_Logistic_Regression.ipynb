# **Logistic Regressio**
👉 **Logistic Regression** predicts probability using sigmoid function and classifies data into classes.

1. Supervised Learning Model
2. Classification model
3. Best for Binary Classification Problem
4. Uses Sigmoid function

## **Sigmoid function**
$$ \hat{Y} = P(\hat{Y}=1|X) = σ(z) = \frac{1}{1 + e^{-z}}$$

* $\hat{Y}$ = probability of $\hat{Y}$ being 1 for the value of X
* $z = w.x + b$  (For Linear regression)
* w = weights (no. of weights is equla to no. of features)
*  b = bias
* if $z < 0$ then $\hat{Y} > 0.5$ and class is 1 using threshold
* if $z > 0$ then $\hat{Y} < 0.5$ and class is 0 using threshold
<br>

## **Advantages:**
    1. Easy to implement
    2. Performs well on data with linear relationship
    3. Less prone to over-fitting for low dimensional dataset

## **Disadvantages:**
    1. High dimensional dataset causes over-fitting
    2. Difficult to capture complex relationships in a dataset
    3. Sensitive to Outliers
    4. Needs a larger dataset

<br>

**Loss Function (L)** → error for one data point

**Cost Function (J)** → average of loss over entire dataset

<br>

## **Logistic regression uses Log Loss / Binary Cross-Entropy**
**Formula(cost function):**
$$J(\theta) = -\frac{1}{m} \sum \left[ \hat{Y} \log(Y) + (1 - \hat{Y})\log(1 - Y) \right]$$

Where:
* m → number of samples
* $Y$ → actual label (0 or 1)
* $\hat{Y}$ → predicted probability

**Formula(Loss function):**
$$J(\theta) = -\left[ \hat{Y} \log(Y) + (1 - \hat{Y})\log(1 - Y) \right]$$

<br>

## **Gradient Descent**
**Formula :**               
$$w = w - L \cdot dw$$
$$b = b - L \cdot db$$

## **Partial derivatives :**
$$dw = \frac{\partial J}{\partial w} = \frac{1}{m} \sum (\hat{Y} - Y)\,X$$

$$db = \frac{\partial J}{\partial b} = \frac{1}{m} \sum (\hat{Y} - Y)$$
