#**🌳 Decision Tree**
A Decision Tree is a supervised ML algorithm used for:
* ✅ Classification
* ✅ Regression

👉 Core idea:

**It splits data into smaller parts using conditions until it reaches a final decision.**

**Example:**

    Is age > 18?
      ├── Yes → Is income > 50k?
      │        ├── Yes → Approve loan ✅
      │        └── No  → Reject ❌
      └── No → Reject ❌

<br>

## **Decision Tree Terminologies**
| Part          | Meaning                   |
| ------------- | ------------------------- |
| Root Node     | Starting point            |
| Decision Node | Condition (feature split) |
| Leaf Node     | Final output              |
| Branch        | Outcome of condition      |

<br>

## **🔢 How Decision Tree Works**

### **Step 1: Choose Best Feature to Split**
👉 The algorithm checks Which **feature best separates** the data?

### **Step 2: Split Data**

**Example:**  age > 30 ?

Data divided into 2 parts

### **Step 3: Repeat Recursively**
* Keep splitting each subset
* Until:
  * Pure data reached
  * OR stopping condition reached

### **Step 4: Make Prediction**
* Traverse tree from root → leaf
* Follow conditions
* **Final leaf = prediction**

<br>

## 🔥**How does it decide “Best Split”?**

### **1. Entropy(Measure of Impurity/randomness)**
**Formula:**
$$Entropy = -Σ_{i=1}^np_i log_2(p_i)$$

**Where:**
* $p_i$ probability of class i
* n is number of classes


**Meaning**:
* High entropy → mixed data(making accurate prediction is tough)
* Low entropy → pure data(making accurate prediction is easier)

<br>

### **2. Information Gain (IG)**
Information Gain (IG) measures the reduction in entropy (uncertainty) after splitting a dataset based on an attribute, defined as the parent entropy minus the weighted average of child entropies.

**Formula:**
$$IG(S, A) = H(S) - Σ_{v𝛜values(A)}\frac{|S_v|}{|S|}H(S_v)$$

**Where:**
* **IG(S, A)** : Information Gain of dataset S by attribute A.
* **H(S)** : Entropy of the parent dataset S before the split.
* $\frac{|S_v|}{|S|}H(S_v)$ : Weighted entropy of the child node $v$ after splitting.

1. **IG tells:** How much impurity is reduced after split
2. Choose feature with **maximum information gain**

<br>

### **3. Gini Index (Alternative to Entropy)**
Gini measures impurity (how mixed the data is)
* If data is pure → Gini = 0
* If data is mixed → Gini is higher

**Formula :**
$$Gini = 1 - 𝜮(p_i)^2$$

<br>

## **👍 Advantages**
* Can be used for both classification and Regression
* Easy to interpret 🌳
* No need for normalization and scaling
* Not sensitive for outliers
* Works with both numeric & categorical data

## **👎 Disadvantages**
* Overfitting : Too many splits → perfect training → poor testing
* Sensitive to small changes
* Can become very deep

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
```

```python
# Load data
data = load_iris()
X, y = data.data, data.target

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = DecisionTreeClassifier(criterion='gini', max_depth=3)

# Train
model.fit(X_train, y_train)
```

```python
# Predict
y_pred = model.predict(X_test)
```

```python
print(y_pred)
print(y_test)
print("Accuracy score : ", accuracy_score(y_test, y_pred)*100)
```

### **⚙️ Important Parameters**
| Parameter           | Meaning              |
| ------------------- | -------------------- |
| `criterion`         | "gini" or "entropy"  |
| `max_depth`         | Maximum tree depth   |
| `min_samples_split` | Min samples to split |
| `min_samples_leaf`  | Min samples in leaf  |
