# **K-Nearest Neighbors (KNN)**

KNN (K-Nearest Neighbors) is a supervised machine learning algorithm and can be used for:
* Classification (e.g., spam or not spam)
* Regression (predicting values)

👉**A data point is classified based on the majority of its nearest neighbors.**

## **How KNN Works (Step-by-Step)**
### **Step 1: Choose K**
* K = number of neighbors
* Example: K = 3, 5, 7

<br>

### **Step 2: Calculate Distance**
Find distance(**Euclidean** or **Manhattan**) between:

* New point (test point) and
* All training points

**Euclidean Distance**
$$d = \sqrt{Σ^n_{i=1}(x_i - y_i)^2}$$

**Manhattan Distance**
$$d = Σ^n_{i=1}|x_i - y_i|$$

**Where**
* $x_i$ = value of feature $i$ of test point
* $y_i$ = value feature $i$ of training point

<br>
  
### **Step 3: Find Nearest Neighbors**
1. Sort distances
2. Pick top K closest points

<br>

### **Step 4: Make Prediction**
#### **For Classification:**
* Take majority vote
* Example:
  * 3 neighbors → (Red, Red, Blue)
  * Output = Red
#### **For Regression:**
* Take average value

<br>

### **Advantages**
* Simple to understand
* No training phase (lazy learner)
* Works well for small datasets with less number of features
* Works for both Classification and Regression
* Easy to implement for Multiclass classification problems

<br>

### **Disadvantages**
* Choosing optimum K value
* Slow for large data (distance calc is heavy)
* Sensitive to Outliers
* Less efficient for high dimensional data
* Does not perform well on imbalanced dataset

### **Euclidean Vs Manhattan**

👉 Use Euclidean:
* जब data smooth हो
* Normal distribution हो

👉 Use Manhattan:
* जब outliers हों
* High dimensional data हो


### **Creating a FUnction for calculating Euclidean DIstance**

```python
import numpy as np
def get_euclidean_distance(p1, p2):
  euclidean_dist = np.sqrt( np.sum( ( np.array(p1) - np.array(p2) )**2 ) )
  print(euclidean_dist)
```

```python
get_euclidean_distance((1,1,1, 6, 5, 3), (2,2,2, 7, 1, 9))
```

```python
get_euclidean_distance((1,1,1), (2,2,2))
```

### **Creating a FUnction for Manhattan DIstance**

```python
import numpy as np
def get_manhattan_distance(p1, p2):
  manhattan_dist = np.sum( abs( np.array(p1) - np.array(p2) ) )
  print(manhattan_dist)
```

```python
get_manhattan_distance((1,1,1,1),(2,2,2,2))
```

```python
print(get_euclidean_distance((63,1,3,145,233,1,0,150,0,2.3,0,0,1),(37,1,2,130,250,0,1,187,0,3.5,0,0,2)))
print(get_manhattan_distance((63,1,3,145,233,1,0,150,0,2.3,0,0,1),(37,1,2,130,250,0,1,187,0,3.5,0,0,2)))
```

**if the distance between two points is very large that means the two points are dissimilar and if the distance between two points is slightly lower that means the two points are similar**
