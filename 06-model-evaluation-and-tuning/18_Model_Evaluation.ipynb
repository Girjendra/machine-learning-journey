# **Evaluation metrics For Classification**

## 1) **Accuracy**
**Accuracy** tells how many total predictions are correct.

**Formula**
$$Accuracy = \frac{TP+TN}{TP+TN+FP+FN​}$$

$$OR$$

$$
\text{Accuracy} = \frac{\text{Number of correct predictions}}{\text{Total number of input data points}}
$$

* **TP (True Positive):** model said **Yes**, and actual answer is also **Yes**
* **TN (True Negative):** model said **No**, and actual answer is also **No**
* **FP (False Positive):** model said **Yes**, but actual answer is **No**
* **FN (False Negative):** model said **No**, but actual answer is **Yes**

## 2) **Precision**
**Precision:** Precision is the ratio of number of True Positive to the total number of Predicted Positive. It measures, out of the total predicted positive, how many are actually positive.

Precision measures the error caused by False Positives. Hence it is a good evaluation metric when False Positive predictions are critical.

**Formula**
$$Precision = \frac{TP}{TP+FP​}$$

## 3) **Recall**
**Recall:** Recall is the ratio of number of **True Positive** to the total number of Actual Positive. It measures, out of the total actual positive, how many are predicted as True Positive.

Recall measures the error caused by **False Negatives**. Hence it is a good evaluation metric when False Negative predictions are critical.

**Formula**
$$Recall = \frac{TP}{TP+FN​}$$

## 4) **F1-score**

**F1-score:** F1 Score is an important evaluation metric for binary classification that combines Precision & Recall. F1 Score is the harmonic mean of Precision & Recall.

This is a very useful metric when a dataset has imbalanced classes.


**Formula**
$$\text{F1-score} = 2×\frac{Precision × Recall}{Precision + Recall​}$$

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# loading the csv data to a Pandas DataFrame
heart_data = pd.read_csv('heart.csv')
```

```python
# print first 5 rows of the dataset
heart_data.head()
```

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

Model Evaluation

**Accuracy Score**

```python
from sklearn.metrics import accuracy_score
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print(training_data_accuracy)
```

```python
print('Accuracy on Training data : ', round(training_data_accuracy*100, 2), '%')
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print(test_data_accuracy)
```

```python
print('Accuracy on Test data : ', round(test_data_accuracy*100, 2), '%')
```

**Precision**

```python
from sklearn.metrics import precision_score
```

```python
# precision for training data predictions
precision_train = precision_score(Y_train, X_train_prediction)
print('Training data Precision =', precision_train)
```

```python
# precision for test data predictions
precision_test = precision_score(Y_test, X_test_prediction)
print('Test data Precision =', precision_test)
```

**Recall**

```python
from sklearn.metrics import recall_score
```

```python
# recall for training data predictions
recall_train = recall_score(Y_train, X_train_prediction)
print('Training data Recall =', recall_train)
```

```python
# recall for test data predictions
recall_test = recall_score(Y_test, X_test_prediction)
print('Test data Recall =', recall_test)
```

**F1 Score**

```python
from sklearn.metrics import f1_score
```

```python
# F1 score for training data predictions
f1_score_train = f1_score(Y_train, X_train_prediction)
print('Training data F1 Score =', f1_score_train)
```

```python
# F1 Score for test data predictions
f1_score_test = recall_score(Y_test, X_test_prediction)
print('Test data F1 Score =', f1_score_test)
```

**Precision, Recall, & F1 Score - function**

```python
def precision_recall_f1_score(true_labels, pred_labels):

  precision_value = precision_score(true_labels, pred_labels)
  recall_value = recall_score(true_labels, pred_labels)
  f1_score_value = f1_score(true_labels, pred_labels)

  print('Precision =',precision_value)
  print('Recall =',recall_value)
  print('F1 Score =',f1_score_value)
```

```python
# classification metrics for training data
precision_recall_f1_score(Y_train, X_train_prediction)
```

```python
# classification metrics for test data
precision_recall_f1_score(Y_test, X_test_prediction)
```

# **Evaluation metrics For Regression**

## 1. **MAE (Mean Absolute Error)**
Average of absolute difference between actual and predicted values.
$$MAE = \frac{1}{n}\sum |y_{true} - y_{pred}|$$

## 2. **MSE (Mean Squared Error)**
Average of squared difference between actual and predicted values.
$$MSE = \frac{1}{n}\sum (y_{true} - y_{pred})^2$$

## 3. **RMSE (Root Mean Squared Error)**
Square root of MSE.
$$RMSE = \sqrt{\frac{1}{n}\sum (y_{true} - y_{pred})^2}$$

## 4. **R² Score**
Shows how well the model explains the variance in the data.
$$R^2 = 1 - \frac{\sum (y_{true} - y_{pred})^2}{\sum (y_{true} - \bar{y})^2}$$

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_true, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R-squared:", r2)
```
