# **Types of Parameters**
| Model Parameters       | Hyperparameters                 |
| ---------------------- | ------------------------------- |
| Learned by model       | Set by user                     |
| Found during training  | Chosen before training          |
| Example: weights, bias | Example: learning rate, k value |

# **Model Parameters**
## **Weights**
**Weights** tell the model **how much influence an input has on the output.**

**If weight is:**
* **high positive** → feature increases output
* **negative** → feature decreases output
* **near zero** → feature has very little effect

<br>

## **Bias**
**Bias** is the offset value given to the **model**. Bias is used to **shift the model** in a particular direction. It is similar to a **Y-intercept**. **bias** is equal to **output** when all the feature values are zero.

$$y = mx + b$$
* **w** = weight
* **x** = input
* **b** = bias

#

# **Hyperparameters**

## **Learning Rate**

**Learning rate** tells the model **how big a step to take while updating weights** during training.

It controls **how fast or slow the model learns.**
* **High learning rate** → big steps, may miss best solution
* **Low learning rate** → small steps, learning is slow

<br>

## **Number of Epochs**

**Epoch** **means one complete pass of the full training dataset by the model.**

**Simple meaning**
* If **epochs** = 10, then model sees the whole training data **10 times**.
