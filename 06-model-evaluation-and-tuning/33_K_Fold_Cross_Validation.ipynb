# **Cross Validation**
Cross validation is used to evaluate the performance of a model(**model selection**) by splitting the dataset multiple times to ensure reliable and unbiased results.

**👉 Problem:**
* Result depends on how data is split
* Different split → different accuracy

So Instead of **testing once** We test the model multiple times on **different splits**


# **K-Fold Cross Validation**

**Split data into K equal parts (folds)**

**Then:**
* Use 1 fold for testing
* Use remaining (K-1) folds for training
* Repeat K times

## **Example (K = 5)**
* Data → divided into 5 parts:
* Fold1 Fold2 Fold3 Fold4 Fold5
| Iteration | Train       | Test |
| --------- | ----------- | ---- |
| 1         | F2 F3 F4 F5 | F1   |
| 2         | F1 F3 F4 F5 | F2   |
| 3         | F1 F2 F4 F5 | F3   |
| 4         | F1 F2 F3 F5 | F4   |
| 5         | F1 F2 F3 F4 | F5   |

## **Final Result**

**👉 Take average accuracy:**

$$Final Score = (Score1 + Score2 + ... + ScoreK) / K$$

**Standard :**
K = 5 or 10

## **Working of K-Fold Cross Validation**

````python
from sklearn.model_selection import KFold
import pandas as pd

kf = KFold(n_splits=5)

data = pd.read_csv('/content/diabetes.csv')

data = []
for train_index, test_index in kf.split(data): # returns indices of train and test data now can use to check performance for any model
  print(train_index, test_index)
````

## **Working of cross_val_score**

````python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

data = pd.read_csv('/content/diabetes.csv')
X = data.drop('Outcome', axis=1)
y = data['Outcome']

model = KNeighborsClassifier(n_neighbors=3) # default is k=5

scores = cross_val_score(model, X, y, cv=5)
print(scores)
````

## **What happens internally:**
* Data split into 5 folds
* Model trains & tests 5 times
* Returns accuracy of each fold

## **Difference Between KFold and cross_val_score**
**1. KFold (Manual Way)**
```bash
kf = KFold(n_splits=5)
for train_index, test_index in kf.split(X):
```
👉 **What it does:**

* Only splits data
* Gives you indices
* You must:
    * Train model manually
    * Predict manually
    * Calculate accuracy manually

* 👉 Control = Full control 🔧
* 👉 Work = More code

## **2. cross_val_score (Automatic Way)**
```bash
scores = cross_val_score(model, X, y, cv=5)
```

**👉 What it does:**
* Splits data ✅
* Trains model ✅
* Tests model ✅
* Returns scores ✅

👉 Control = Less control <br>
👉 Work = Very easy 🚀
