# **Supervised Learning**

**Supervised Learning** is the learning where We give the model **input + correct output**, and it learns the **relationship**

## **Types of Supervised Learning** :

1️⃣ **Classification**

Output is **categorical** (labels)

👉 **Example**:

Email → Spam or Not Spam

Disease → Positive or Negative
<br><br>
2️⃣ **Regression**

Output is **continuous** (numbers)

👉 **Example**:

House price prediction

Marks prediction
<br>
## **Classification Models** :

* Logistic Regression

* Support Vector Machine Classifier

* Decision Tree

* K-Nearest Neighbors

* Random Forest

* Naive Bayes Classifier

<br>

## **Regression Models**

* Linear Regression

* Lasso Regression

* Polynomial Regression

* Support Vector Machine Regressor

* Random Forest Regressor

* Bayesian Linear Regressor

# **Unsupervised Learning**

**Unsupervised Learning** is a learning where We give only **input** data (no answers), and the model finds hidden **patterns** or **groups**.

## **Types of Unsupervised Learning**

1️⃣ **Clustering**

Groups **similar** data points together

**Example**: customer segmentation

2️⃣ **Association**

Finds **relationships** between items

**Example**: “People who buy milk also buy bread”

3️⃣ **Dimensionality Reduction**

Reduces number of features

**Example**: PCA

## **Unsupervised Learning Models**

* K-Means Clustering

* Hierarchical Clustering

* Principal Component Analysis (PCA)

* Apriori

* Eclat
