# **Machine Learning Model**

A **Machine Learning model** is a **program** or **mathematical function** that learns patterns from data and makes **predictions** or **decisions**.

At the very beginning, there is no learned function yet. What we have is only a **model structure** (**hypothesis**).

**Step-by-step**

1️⃣ **Initially** :

We assume a form of function.

Example (**Linear Regression**):
$$y = w_1x_1 + w_2x_2 + b$$
But the parameters w1, w2 and b are **unknown**.

2️⃣ **Training phase** :

The algorithm uses data to learn the parameters.

It adjusts w1, w2 and b to reduce prediction error.

3️⃣ **After training** :

Now the parameters have values.

**Example**:
$$y = 2.3x_1 + 1.1x_2 + 5$$

Now we have a final learned function → this is the ML model.

Before training

➡️ **Empty formula (unknown parameters)**

After training

➡️ **Learned mathematical function**

So Initially there is only a **model structure**, and after training it becomes a **learned mathematical function (ML model).**

**How a ML Model Works**

1️⃣ **Training**

Model learns patterns from data.

2️⃣ **Learning Pattern**

Finds relationship between input and output.

3️⃣ **Prediction**

Uses the learned pattern to predict new results.

**Simple Example of ML Models**

* Linear Regression

* Decision Tree

* Neural Network

* Random Forest

* Support Vector Machine (SVM)
