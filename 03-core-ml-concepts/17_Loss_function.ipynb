# **Loss function**
It measures the **difference** between **actual** answer and **predicted** answer.

**Smaller loss = better model**

## **Types of Loss Function**
1. **Squared Error Loss(MSE)** : Used in **regression**
  $$MSE = \frac{1}{n}Σ(y - ŷ)^2$$
    y = **Actual** value

    $ŷ$ = **Pridicted** value

2. **Cross Entropy Loss** : Used  in **classification** problems.

    It checks how wrong the predicted class **probability** is.

3. **KL Divergence** : It measures how one **probability distribution** is different from another probability distribution.

<br>

* **Loss high** → model bad
* **Loss low** → model good
