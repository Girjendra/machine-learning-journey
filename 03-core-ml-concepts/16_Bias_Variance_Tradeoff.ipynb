# 1️⃣ **Bias**
**Bias** is the **difference** between the **predicted value** of our model and the **correct value** which we are trying to predict.

<br>

# 2️⃣ **Variance**
**Variance** tells us how much the **model’s prediction changes** when we train it on **different datasets**.

<br>

(i) **Underfitting**
1. High Bias
2. Low Variance

(ii) **Overfitting**
1. Low Bias
2. High Variance

<br>

# **Bias–Variance Tradeoff**

The **Bias–Variance** Tradeoff is the balance between:
1. **Bias** → error due to too simple model
2. **Variance** → error due to too complex model

👉 **In simple words:** We try to find a model that is not **too simple** and not **too complex**.

<br>

# **Techniques to have better Bias – Variance Tradeoff:**

1. **Good Model Selection** :
Choosing the right model complexity for the problem so it neither underfits nor overfits.

2. **Regularization** :
Adding a penalty to the model to reduce overfitting and keep the model simpler.

3. **Dimensionality Reduction** :
Reducing the number of input features to remove unnecessary data and improve model performance.

4. **Ensemble methods** :
Combining multiple models to get better and more stable predictions.
