# **Overfitting**
**Overfitting** happens when a model **learns the training data too well**, including noise and small details, and then **performs poorly on new data**.

<br>

## **How to identify**
**Training accuracy** → very high ✅

**Testing accuracy** → low ❌

<br>

## **Causes for Overfitting:**
1. Less Data
2. Increased Complexity of the model
3. More number of layers in Neural Network
4. Too many features

<br>

## **Preventing Overfitting by:**
1. Using more data
2. Reduce the number of layers in the Neural network
3. Early Stopping
4. Bias – Variance Tradeoff
5. Use Dropouts
6. Remove unnecessary features

# **Underfitting**
**Underfitting** is a condition where a model **fails to learn the underlying patterns in the data**, resulting in **poor performance** on both **training** and **test** data.

<br>

## **How to identify**
**Training accuracy** → low ❌

**Testing accuracy** → low ❌

<br>

## **Causes for Underfitting:**
1. Choosing a wrong model
2. Less complexity of the model
3. Less variance but high bias

<BR>

## **Prevent Underfitting by:**
1. Choosing the correct model appropriate for the problem
2. Increasing the complexity of the model
3. More number of parameters to the model
4. Bias – Variance Tradeoff

# **Regularization**

**Regularization** is a technique used to **reduce overfitting** by **penalizing large model parameters (weights)**.

<br>

## **Formula idea**
$$Loss = Error + λ*Penalty$$
λ = controls strength of regularization

<br>

## **Types**
1. **L1 Regularization (Lasso)** → makes some weights = 0 (feature selection)
2. **L2 Regularization (Ridge)** → reduces weights but not zero
