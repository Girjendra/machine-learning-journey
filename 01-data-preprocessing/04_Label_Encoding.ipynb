**Label Encoding** is a technique used to convert categorical (text) data into numerical form so that machine learning models can understand it.

👉 ML models cannot work directly with text like "Male", "Female".

🔹 Where Label Encoding is Used

✅ Target column (very common)

```python
import sklearn.datasets
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
print(dir(sklearn.datasets))

idf = sklearn.datasets.load_iris()
bcd = sklearn.datasets.load_breast_cancer()
```

```python
id = pd.DataFrame(idf.data, columns=idf.feature_names)
cd = pd.DataFrame(bcd.data, columns=bcd.feature_names)

# print(id.head())
# print(cd.head())
```

```python
col = bcd['target']
col2 = []

for i in range(len(col)):
    if col[i] == 0:
        col2.append('M')
    else:
      col2.append('B')

cd['daignosis'] = col2
print(cd.head())
print(cd['daignosis'].value_counts())
```

```python
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(cd.daignosis)
cd.daignosis = labels
print(cd.daignosis.value_counts())
print(cd.head())
```

```python
col = idf.target
col2 = []

for i in range(len(idf.target)):
  if col[i] == 0:
    col2.append('setosa')
  elif col[i] == 1:
    col2.append('versicolor')
  else:
    col2.append('virginica')

id['species'] = col2
print(id.species.value_counts())
print(id.head())
```

```python
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(id.species)
id.species = labels
print(id.species.value_counts())
print(id.head())
```
