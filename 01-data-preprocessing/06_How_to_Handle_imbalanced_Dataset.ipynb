### An imbalanced dataset occurs when the number of samples in each class is not equal.

```python
df = pd.read_csv("/content/credit_data.csv")
df = pd.DataFrame(df)
print(df.Class.value_counts())
```

```python
wrong = df[df.Class == 1]
correct = df[df.Class == 0]
```

```python
print(wrong.shape)
print(correct.shape)
```

```python
correct_sample = correct.sample(n=213)
print(correct_sample.shape)
```

```python
new_df = pd.concat([wrong, correct_sample], axis = 0)
print(new_df.shape)
```
