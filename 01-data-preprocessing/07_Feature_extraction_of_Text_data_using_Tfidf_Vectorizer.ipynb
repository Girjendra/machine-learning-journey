Machine learning models cannot understand text directly.
So we convert text into numerical features (vectors).

This process is called Feature Extraction.

TF-IDF (Term Frequency – Inverse Document Frequency) converts text into numerical vectors based on word importance.

👉 It gives higher weight to important words and lower weight to common words.

1️⃣ Term Frequency (TF)

      TF = word count in document / total words in document

2️⃣ Inverse Document Frequency (IDF)

      IDF=log( Total Documents / Documents containing the word )


3️⃣ Final TF-IDF

        TF-IDF=TF×IDF

👉 Important words get high score.


Bag of Words (BoW) is a technique used in Natural Language Processing (NLP) to convert text data into numerical form by counting how many times each word appears.

corpus means collection of words.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = [
    "I love machine learning",
    "Machine learning is amazing",
    "I love AI"
]

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(corpus)

print(X)
```
