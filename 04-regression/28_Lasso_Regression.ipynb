# **Lasso Regression**
Lasso Regression is a type of linear regression that adds a penalty on absolute values of weights (L1).

### **It reduces overfitting and can remove unnecessary features**

## **📊 Formula:**
$$J=MSE+λ∑|w|$$

$$MSE = \frac{1}{m}​∑(y−\hat{y}​)2$$
where:
* MSE → error
* |w| → absolute weights
* λ → regularization streangth
* $y$ → actual value
* $\hat{y}$→ predicted value

📌 **Behavior:**
* λ = 0 → normal linear regression
* λ ↑ → more weights = 0
* λ बहुत बड़ा → almost sab remove

👉 Lasso regression uses L1 regularization to shrink some weights to zero, performing feature selection.

* **LASSO = Least Absolute Shrinkage and Selection Operator**

<br>

## **Gradient Descent for LASSSO Regression**
Gradient Descent is an optimization algorithm used for minimizing the cost function in various machine learning algorithms. It is used for updating the parameters of the learning model.

$$w = w - L \cdot \frac{\partial J}{\partial w}$$

$$b = b - L \cdot \frac{\partial J}{\partial b}$$

**where :**

* $w \rightarrow \text{weight}$
* $b \rightarrow \text{bias}$
* $L \rightarrow \text{Learning Rate}$
* $\frac{\partial J}{\partial w} \rightarrow \text{Partial Derivative of cost function w.r.t } w$
* $\frac{\partial J}{\partial b} \rightarrow \text{Partial Derivative of cost function w.r.t } b$
