**Matrix** : A matrix is a rectangular table of numbers arranged in rows and columns.

```python
import numpy as np
```

```python
A = np.array([[1,2,3],[4,5,6]])

print(A)
print(A.shape)
```

```python
zero_matrix = np.zeros((3,3))
print(zero_matrix)
```

```python
identity_matrix = np.eye(3)
print(identity_matrix)
```

```python
random_matrix = np.random.rand(3,3)
print(random_matrix)
```

```python
ones_matrix = np.ones((3,3))
print(ones_matrix)
```

**Transpose of a matrix** : The transpose of a matrix means rows become columns and columns become rows.
$$A^{T}$$

```python
import numpy as np

A = np.array([[1,2,3],[4,5,6]])
print(A)

# transpose_matrix = A.T
transpose_matrix = np.transpose(A)
print(transpose_matrix)
```

```python
A = np.array([[1,2],[3,4]])

B = np.array([[5,6],[7,8]])

# C = A + B
C = np.add(A, B)

print("Matrix Addition:")
print(C)
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

# D = A - B
D = np.subtract(A, B)

print("Matrix Subtraction:")
print(D)
```

```python
import numpy as np

A = np.array([[1,2],[3,4]])

B = np.array([[5,6],[7,8]])

# C = np.multiply(A, B)
C = A * B
print("Matrix Multiplication(elementwise):")
print(C)
```

```python
A = np.array([[1,2],[3,4]])
result = 3 * A

print("Scalar Multiplication:")
print(result)
```

```python
import numpy as np

A = np.array([[1,2],[3,4]])

B = np.array([[5,6],[7,8]])

# C = np.matmul(A, B)
C = A @ B
print("Matrix Multiplication:")
print(C)
```

$$dodproduct = multiplication$$

```python
import numpy as np

A = np.array([[1,2],[3,4]])

B = np.array([[5,6],[7,8]])

# C = np.matmul(A, B)
C = np.dot(A, B)
print("Dod product:")
print(C)
```
