# **Probability**
Probability is a numerical measure of the likelihood that an event will occur.

## **Range of Probability**
**0**  → impossible event

**1**  → certain event

**Simple Formula**

$$ Probability = \frac{Number of favorable outcomes}{Total outcomes}$$
	​



## **Random Variable**
A random variable is a variable whose value depends on the outcome of a random experiment.

## **Types of Random Variables**
1️⃣ **Discrete Random Variable**

A discrete random variable takes only countable or descrete values.

**Examples** :

* Number of heads in 3 coin tosses

* Number of students in a class

2️⃣ **Continuous Random Variable**

A continuous random variable can take any value in a range.

**Examples**  :

* Height of students

* Weight of a person

## **Probability Distribution**

A probability distribution describes how probabilities are assigned to the possible values of a random variable.

**Example (Discrete Random Variable)**

Suppose we toss a **coin 2 times.**

Let **X = number of heads**

Possible values of X: 0, 1, 2

Probability distribution:
| X (Heads) | Probability P(X) |
| --------- | ---------------- |
| 0         | 1/4              |
| 1         | 1/2              |
| 2         | 1/4              |


This table is the **probability distribution.**

## **Types of Probability Distribution**
1️⃣ **Discrete Probability Distribution** :

Used when the random variable has countable values.

2️⃣ **Continuous Probability Distribution**

Used when the random variable can take any value in a range.



## **Normal Distribution**

A Normal Distribution(**Gaussian distribution**) is a distribution where data is symmetrically distributed around the mean.

It forms a **bell-shaped curve**.

**Key Properties**

* **Mean = Median = Mode**

* Data is **symmetrical**

* Most values lie near the **mean**

* Fewer values are at the extremes

**Example:**

Heights of people in a population often follow a normal distribution.

## **Skewness**
A data distribution is considered skewed when the distribution curve appears distorted or shifted to the left or right in a statistical distribution.

**Types of Skewness**

1️⃣ **Positive Skew (Right Skew)** :

* Tail is longer on the right

* Mean > Median > Mode

2️⃣ **Negative Skew (Left Skew)**

* Tail is longer on the left

* Mean < Median < Mode

## **Poisson Distribution**

**Poisson Distribution** is a probability distribution that measures how many times an event is likely to occur within a specified period of time.

**Examples** :

* Number of calls received in a call center per hour

* Number of cars arriving at a traffic signal per minute

**Poisson distribution is used when:**

* Events occur randomly

* Events occur independently

* We count number of occurrences

**Formula** :

$$P(x) = \frac{e^{-λ}λ^{x}}{x!}$$
**Where:**

λ = average number of events

x = number of events

e ≈ 2.718 (Euler's number)
