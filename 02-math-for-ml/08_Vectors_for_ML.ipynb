```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Vector** : A vector in machine learning is a list of numerical features representing a data point that algorithms can process mathematically.

**Feature Vector** :
When all features of a sample are written as numbers together, it is called a **feature** vector.

➕ **Vector Addition** :

Add the corresponding elements of the vectors.
$$A + B = (a_{1}+b_{1}, a_{2}+b_{2}, a_{3}+b_{3})$$

➖ **Vector Subtraction** :

Subtract corresponding elements.
$$A + B = (a_{1}-b_{1}, a_{2}-b_{2}, a_{3}-b_{3})$$

**Angle of a vector from x axis** :
$$θ = tan^{-1}\frac{y}{x}$$

```python
A = np.array([2,3])
B = np.array([4,1])

print("addition : ", A + B)   # addition
print("subtraction : ", A - B)   # subtraction
print("scalar multiplication : ", 2*A)   # scalar multiplication
```

**Dod product of to vectors** :
$$\vec{a}.\vec{b} = |\vec{a}||\vec{b}|\cosθ$$

**Cross product of to vectors** :
$$\vec{a}X\vec{b} = |\vec{a}||\vec{b}|\sinθ\hat{n}$$

**Projection of vector a on vector b** :
$$\text{proj}_{\vec{b}}\vec{a} = \frac{\vec{a}\cdot\vec{b}}{|\vec{b}|^2}\vec{b}$$

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

dot_product = np.dot(a, b)
print("dot_product : ", dot_product)  # 32
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

cross_product = np.cross(a, b)
print("cross_product : ", cross_product)  # [-3, 6, -3]
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

projection = (np.dot(a, b) / np.dot(b, b)) * b

print("projection of a on b: ", projection) #
```
